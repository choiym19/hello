# Factors Affecting Pay Differences

```python
import pandas as pd
import seaborn as sns
```

```python
url = 'http://bit.ly/hr_dataset'
df_employees = pd.read_csv(url)
```

```python
df_employees.shape
```

```python
df_employees.describe()
```

From the describe function we get employee number, age and pay rate, this is because they are intergers. We do not learn much from the description of employee number since these are unique identifiers. For age, we learn that mean age is 38.87 and for pay rate it is $31.28.

```python
df_employees.head()
```

```python
#looking at distribution of wage
sns.displot(df_employees, x= "Pay Rate", kind ='kde', fill = True)
```

This histogram shows the pay rate histogram. It seems that there are two clear common wages. One is at $20 and one is around $60. We can describe the shape as being a bimodial distribution. It would make sense that these two peak exists becuase of jobs of different skill and/or experience level. Not surprisingly, the lower wage job has a higher distribution.

```python
#comparing male and female for wage distribution
sns.displot(df_employees, x = "Pay Rate", hue="Sex", kind = "kde", fill = True)
```

This graph looks at the difference of wage distribution based on gender. The initial read on the trend is somewhat worrisome. It seems that there are more males at the higher wage peak (around $60) and less males at the lower wage peak (around $20). From the graph, it is hard to say that the difference at the higher wage peak is significant. However, the density of females is clearly larger than males in the lower wage area. This would suggest that many more female employees earn a lower wage than their male counterparts. This may be sign of wage disparities between male and females, however further analysis would be necessary to know for sure.

```python
#comparing race for wage distribution
sns.displot(df_employees, x = "Pay Rate", hue="RaceDesc", kind = "kde", fill = True)
```

This graphs looks at the difference of wage distribution between different recorded races. The first thing to note is that there seems to be a significantly higher number of White employees than other employees. Also, it seems that the peaks for these white employees are higher for the lower and higher wage peak. The next highest population seems to be African American emloyees. In genreal, it seems that there is a similar trend of more people in the lower wage bracket and less people in the higher wage bracket across races.

```python
#Distribution 
sns.displot(df_employees, x = "Pay Rate", hue="Employment Status", kind = "kde", fill = True)
```

From the above chart, we can see the distribution of wages for employment status. For people who are actively employed, it seems that the distribution of pay is similar to what we observed initially. Looking at people who were voluntarily terminated it seems that the majority of the people were those in the lower wage bracket. It seems that the number of people in terminates for cause, on the other hand, have a more even balance between the lower wage bracket and higher wage bracket. However, the number of people in this group is quite low and it is difficult to confidently conclude that this is the true trend for that group. For voluntarily terminated, it seems that the trend is more obvious and significant.
